```python
import numpy
```

```python
candidate_points = {
    "iris_sertosa": numpy.array([5.8, 4.0, 1.2, 0.2]),
    "iris_versicolor": numpy.array([6.1, 2.8, 4.0, 1.3]),
    "iris_virginica": numpy.array([6.3, 2.7, 4.9, 1.8]),
}
```

```python
def find_candidate(point: numpy.ndarray) -> str:
    distance1: float = numpy.dot(point.T, candidate_points["iris_sertosa"])
    distance2: float = numpy.dot(point.T, candidate_points["iris_versicolor"])
    distance3: float = numpy.dot(point.T, candidate_points["iris_virginica"])

    min_distance: float = min([distance1, distance2, distance3])
    print(numpy.array([distance1, distance2, distance3, min_distance]))
    if distance1 == min_distance:
        return "iris_sertosa"
    elif distance2 == min_distance:
        return "iris_versicolor"
    elif distance3 == min_distance:
        return "iris_virginica"
    else:
        raise Exception()

    
```

```python
with open("iris.data") as file:
    for line in file:
        line = line.strip().lower()
        numbers = list(map(float, line.split(",")[:-1]))
        point = numpy.array(numbers)

        print()
        
        print(find_candidate(point), end= "\t")
        print(*list(point), sep = ",")
```
