```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python

datafile = "adult.csv"
```

```python

# read list of words from file column_names.txt
column_names = open("column_names.txt", "r").read().splitlines()
```

```python
column_names
```

```python

# read data from file adult.csv
data = pd.read_csv(datafile, names=column_names, header=None, index_col=False, sep=",")
```

```python
data
```

```python
data["workclass"]
```

```python
X = data[['age', 'hours-per-week', 'capital-loss']].copy()
X
```

```python
# use KMeans clustering on data
kmeans = KMeans(n_clusters=10).fit(data[['age', 'hours-per-week', 'capital-loss']])
```
